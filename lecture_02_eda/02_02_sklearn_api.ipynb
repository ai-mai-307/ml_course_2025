## Что такое sklearn transformers?

```python
class ExampleDataTransformer():
    def fit(self, X, y=None):
        # Узнать параметры для трансформации
        return self

    def transform(self, X):
        # Применить преобразование к данным
        return X
    
    def fit_transform(self, X, y=None):
        self.fit(X, y)
        return self.transform(X)
```

Как в таком случае будет выглядеть реализация min-max масштабирования?

```python
class CustomMinMaxScaler:
    def __init__(self):
        self.min_value = None
        self.max_value = None

    def fit(self, X, y=None):
        self.max_value = X.max()
        self.min_value = X.min()
        return self
    
    def transform(self, X):
        if self.min_value is None or self.max_value is None:
            raise ValueError("Неизвестно минимальное или максимальное значение.")
        return (X - self.min_value) / (self.max_value - self.min_value)
    
    def fit_transform(self, X, y=None):
        return self.fit(X, y).transform(X)
```

```python
minmaxscaler = CustomMinMaxScaler()
minmaxscaler
```

### Объединяем пайплайны

Существует несколько способов в рамках одного пайплайн реализовать разветвленную обработку данных, это ColumnTransforfmer и FeatureUnion. Рассмотрим их поочередно

#### FeatureUnion

```python
from sklearn.pipeline import Pipeline, FeatureUnion

feature_union = FeatureUnion([

])
```

#### ColumnTransformer

```python
simple_imputer = SimpleImputer(strategy='median')
std_scaler = StandardScaler()

pipe_num = Pipeline([('imputer', simple_imputer), ('scaler', std_scaler)])

s_imputer = SimpleImputer(strategy='constant', fill_value='unknown')
ohe_encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
pipe_cat = Pipeline([('imputer', s_imputer), ('encoder', ohe_encoder)])

col_transformer = ColumnTransformer([('num_preproc', pipe_num, [x for x in df.columns if df[x].dtype!='object']),
                                     ('cat_preproc', pipe_cat, [x for x in df.columns if df[x].dtype=='object'])])

final_pipe = Pipeline([('preproc', col_transformer)])
```
